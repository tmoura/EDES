# BubbleSort

```python
def bubbleSort(tabela):
    n = len(tabela)
    # Percorre todos os elementos da lista
    for i in range(n):
        for j in range(0, n-1):
            if tabela[j] > tabela[j+1]:
                aux = tabela[j]
                tabela[j] = tabela[j+1]
                tabela[j+1] = aux
```

# BubbleSort_melhor

```python
def bubbleSort_melhor(tabela):
   ordenado = 0

   while not ordenado:
      ordenado = 1  # Supõe que a tabela está ordenada

      for i in range(len(tabela) - 1):

        if (tabela[i] > tabela[i+1]):  # Compara elementos adjacentes
          ordenado = 0  # Pelo menos um elemento está fora de ordem

          # Troca elementos adjacentes
          aux = tabela[i]
          tabela[i] = tabela[i+1]
          tabela[i+1] = aux
```

# QuickSort

```python
def particiona(tabela, inf, sup):

   # pivo - Elemento cuja posição final é procurada
   # Por simplicidade, o pivô é o primeiro item da tabela

   pivo = tabela[inf]
   sobe = inf  # Começa no início da tabela e segue em direção do final
   desce = sup # Começa no final da tabela e segue em direção do início

   while (sobe < desce):
      while (tabela[sobe] <= pivo and sobe < sup):  # Vai subindo até o item
         sobe += 1        # apontado por 'sobe' tornar-se maior do que o pivô

      while (tabela[desce] > pivo): # Vai descendo até o item apontado por
         desce -= 1   	# 'desce' tornar-se menor do que ou igual ao pivô

      if (sobe < desce): # Troca as posições de tabela[desce] e tabela[sobe]
         aux = tabela[desce]
         tabela[desce] = tabela[sobe]
         tabela[sobe] = aux


   # Troca as posições de tabela[desce] com tabela[inf]
   tabela[inf] = tabela[desce]
   tabela[desce] = pivo

   return desce # 'desce' está apontando para a posição final do pivô


def quick(tabela,inf,sup):

  if (inf < sup): # A tabela ainda não está ordenada
    posPivo = particiona(tabela, inf, sup) # Faz a partição da tabela
    quick(tabela, inf, posPivo-1) # Ordena primeira partição da tabela
    quick(tabela, posPivo+1, sup) # Ordena segunda partição da tabela


def quickSort(tabela):

   quick(tabela, 0, len(tabela) - 1)
```

# SelectionSort

```python
def selectionSort(tabela):

   for i in range(len(tabela)-1,1,-1):
      maior = tabela[0]
      indice = 0

      for j in range(1,i+1):
        if (tabela[j] > maior):
          maior = tabela[j]
          indice = j

      tabela[indice] = tabela[i]
      tabela[i] = maior


def selectionSort_2(tabela):

   for i in range(len(tabela)):
      maior = tabela[i]
      indice = i

      for j in range(i+1,len(tabela)):
        if (tabela[j] < maior):
          maior = tabela[j]
          indice = j

      tabela[indice] = tabela[i]
      tabela[i] = maior
```

# ShellSort

```python
def shellSort(tabela):
    count = len(tabela)//2

    while count > 0:

      for start in range(count):
        for i in range(start+count,len(tabela),count):

            value = tabela[i]
            position = i

            while position>=count and tabela[position-count]>value:
                tabela[position] = tabela[position-count]
                position = position-count

            tabela[position] = value
      count = count // 2
```

# MergeSort

```python
def Merge(tabela,aux,iniEsq,iniDir,finalDir):
  finalEsq = iniDir - 1
  posAux = iniEsq
  nElementos = finalDir - iniEsq + 1

  while((iniEsq <= finalEsq ) and (iniDir <= finalDir)): # Faz a intercalação
    if(tabela[iniEsq] <= tabela[iniDir]):           # das duas metades
      aux[posAux] = tabela[iniEsq]			# colocando o resultado
      posAux += 1						# na tabela auxiliar
      iniEsq += 1
    else:
      aux[posAux] = tabela[iniDir]
      posAux += 1
      iniDir += 1

  while(iniEsq <= finalEsq):  # Copia o restante da primeira metade
    aux[posAux] = tabela[iniEsq]
    posAux += 1
    iniEsq += 1

  while(iniDir <= finalDir): # Copia o restante da segunda metade
    aux[posAux] = tabela[iniDir]
    posAux += 1
    iniDir += 1

  for i in range(nElementos):  		# Copia a tabela auxiliar
    tabela[finalDir] = aux[finalDir]	# de volta na tabela original
    finalDir -= 1


def MSort(tabela,aux,esquerda,direita):

   if(esquerda < direita):
      centro = (esquerda + direita)//2

      MSort(tabela, aux, esquerda, centro) # Ordena metade inferior da tabela
      MSort(tabela, aux, centro+1, direita) # Ordena metade superior da tabela

      	# Intercala as duas metades da tabela na tabela auxiliar
      Merge(tabela, aux, esquerda, centro+1, direita)


def MergeSort(tabela):

   # Aloca espaço para a tabela auxiliar

   aux = [None] * len(tabela) # Cria aux com tamanho igual ao tamanho da tabela

   MSort(tabela, aux, 0, len(tabela) -1) # Chama MSort() para fazer o serviço

   aux = None # Libera o espaço ocupado pela tabela auxiliar
```
