# Questão 1



```python
chaves = [10, 15, 18, 23, 25, 30, 32, 38, 45, 47, 50, 63, 67, 72, 80, 95]

def busca_sequencial(chaves, k):
  for i in range(len(chaves)):
    if chaves[i] == k:
      return i
    else:
      if k < chaves[i]:
        return -1
  return -1

index = busca_sequencial(chaves, 24)

print(index)
```

# Questão 2

```python
chaves = [10, 15, 18, 23, 25, 30, 32, 38, 45, 47, 50, 63, 67, 72, 80, 95]

chaves.reverse()

print(chaves)

def busca_binaria(chaves, k):
  inf = 0
  sup = len(chaves) - 1

  while inf <= sup:
    meio = (inf + sup) // 2
    if chaves[meio] == k:
      return meio
    elif chaves[meio] > k:  # a diferença tá aqui
      inf = meio + 1
    else:
      sup = meio - 1
  return -1

index = busca_binaria(chaves, 25)

print(index)
```

# Questão 3

```python
chaves = [10, 15, 18, 23, 25, 30, 32, 38, 45, 47, 50, 63, 67, 72, 80, 95]

chaves.reverse()

print(chaves)

def busca_binaria_rec(chaves, k, inf, sup):

  if inf <= sup:
    meio = (inf + sup) // 2

    if chaves[meio] == k:
      return meio
    elif chaves[meio] > k:
      return busca_binaria_rec(chaves, k, meio+1, sup)
    else:
      return busca_binaria_rec(chaves, k, inf, meio-1)
  return -1

index = busca_binaria_rec(chaves, 25, 0, len(chaves) - 1)

print(index)
```

# Questão 4

```python
chaves = [i * 10 for i in range(1024)]

print(chaves)

def busca_binaria_normal(chaves, k):
  inf = 0
  sup = len(chaves) - 1

  while inf <= sup:
    meio = (inf + sup) // 2
    if chaves[meio] == k:
      return meio
    elif chaves[meio] < k: # mudei aqui
      inf = meio + 1
    else:
      sup = meio - 1
  return -1


def busca_binaria_interpolacao(chaves, k):
  inf = 0
  sup = len(chaves) - 1

  while inf <= sup:
    meio = int(inf + (sup - inf) * ((k - chaves[inf])/(chaves[sup] - chaves[inf])))

    if chaves[meio] == k:
      return meio
    elif chaves[meio] < k: # mudei aqui
      inf = meio + 1
    else:
      sup = meio - 1
  return -1


index = busca_binaria_interpolacao(chaves, 500)

print(index)
```
