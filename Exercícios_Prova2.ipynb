```python
chaves = [10, 15, 18, 23, 25, 30, 32, 38, 45, 47, 508, 63, 67, 72, 80, 95]

def busca_sequencial(chaves, k):
  for i in range(len(chaves)):
    if chaves[i] == k:
      return i
    else:
      if k < chaves[i]:
        return -1
  return -1

index = busca_sequencial(chaves, 24)

print(index)
```
