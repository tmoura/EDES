# Gerar Vetor de Chaves

```python
import random

# Gera um número inteiro aleatório entre 1 e 100 (inclusive)
chaves = []

for i in range(1,20):
  chaves.append(random.randint(1, 100))

print(chaves)
```

# Busca Sequencial

```python
def busca_sequencial(chaves, chave_procurada):
  for i in range(1, len(chaves)):
    if chaves[i] == chave_procurada:
      return i
  return -1


## Testando
chave_procurada = 20
indice = busca_sequencial(chaves, chave_procurada)

print(indice)
```

# Busca Binária

```python
chaves = sorted(chaves)
print(chaves)

def busca_binaria(chaves, chave_procurada):
  inf = 0
  sup = len(chaves) - 1

  while inf <= sup:
    meio = (inf + sup) // 2
    if chaves[meio] == chave_procurada:
      return meio
    elif chaves[meio] < chave_procurada:
      inf = meio + 1
    else:
      sup = meio - 1
  return -1


## Testando
chave_procurada = 20
indice = busca_binaria(chaves, chave_procurada)

print(indice)
```

# Busca por Interpolação

```python
chaves = [5,10,15,20,25,30,35,40,45,50,55,60,65,70]

def busca_interpolacao(chaves, chave_procurada):
  inf = 0
  sup = len(chaves) - 1

  while inf <= sup:
    meio = meio = inf + (sup-inf) * ((chave_procurada - chaves[inf]) / (chaves[sup] - chaves[inf]))
    if chaves[meio] == chave_procurada:
      return meio
    elif chaves[meio] < chave_procurada:
      inf = meio + 1
    else:
      sup = meio - 1
  return -1


## Testando
chave_procurada = 20
indice = busca_binaria(chaves, chave_procurada)

print(indice)
```
