```python
class Indice_Chave:
  def __init__(self):
    self.__chave = 0
    self.__indice = 0

    @property
    def chave(self):
      return self.__chave

    @property
    def indice(self):
      return self.__indice

    @chave.setter
    def chave(self, chave):
      self.__chave = chave

    @indice.setter
    def indice(self, indice):
      self.__indice = indice
```

```python
class No:
  def __init__(self, chaves, N):
    self._chaves = [None] * N-1   # Lista de elementos do tipo: IndiceChave
    self._filhos = [None] * N     # Lista de elementos do tipo: No
    self._nFilhos = len(chaves) + 1

    for i in range(len(chaves)):
      item = Indice_Chave()
      item.set_chave(chaves[i])
      item.set_indice(i)

      self._chaves[i] = item

  @property
  def chaves(self):
    return self._chaves

  @property
  def filhos(self):
    return self._filhos

  @property
  def nFilhos(self):
    return self._nFilhos

  @chaves.setter
  def chaves(self, chaves):
    self._chaves = chaves

  @filhos.setter
  def filhos(self, filhos):
    self._filhos = filhos

  @nFilhos.setter
  def nFilhos(self, nFilhos):
    self._nFilhos = nFilhos
```

```python
class Arvore_Multi:

  def __init__(self):
    self.__raiz = None

  @property
  def raiz(self):
    return self.__raiz

  @raiz.setter
  def raiz(self, raiz):
    self.__raiz = raiz

  def busca_em_no(self, chave, no):
    n_Chaves = no.nFilhos - 1

    for i in range(n_Chaves):
      if (chave <= no.chaves[i].chave):
        return i

    return n_Chaves

  def busca_multi(self, chave):
    arvore = self.__raiz

    while (arvore != None):
      i = self.busca_em_no(chave, arvore)

      itemChave = arvore.chaves[i].chave

      if((i < arvore.nFilhos - 1) and (chave == itemChave)):
        return arvore.chaves[i].indice

      arvore = arvore.filhos[i]

      return -1

  def caminhamento(self):
    if (self.__raiz != None):
      for i in range(self.__raiz.nFilhos - 1):
        self.caminhamento(self.__raiz.filhos[i])
        print(self.__raiz.chaves[i].chave)

      self.caminhamento(self.__raiz.filhos[self.__raiz.nFilhos - 1])
```
