```python
class Indice_Chave:
  def __init__(self):
    self.__chave = 0
    self.__indice = 0

    @property
    def chave(self):
      return self.__chave

    @property
    def indice(self):
      return self.__indice

    @chave.setter
    def chave(self, chave):
      self.__chave = chave

    @indice.setter
    def indice(self, indice):
      self.__indice = indice
```

```python
class No:

  # Ordem da √Årvore
  ordem = 4

  def __init__(self, chaves):
    self._chaves = [None] * (No.ordem - 1)   # Lista de elementos do tipo: Indice_Chave
    self._filhos = [None] * No.ordem     # Lista de elementos do tipo: No
    self._nFilhos = len(chaves) + 1

    for i in range(len(chaves)):
      item = Indice_Chave()
      item.chave = chaves[i]
      item.indice = i

      self._chaves[i] = item

  @property
  def chaves(self):
    return self._chaves

  @property
  def filhos(self):
    return self._filhos

  @property
  def nFilhos(self):
    return self._nFilhos

  @chaves.setter
  def chaves(self, chaves):
    self._chaves = chaves

  @filhos.setter
  def filhos(self, filhos):
    self._filhos = filhos

  @nFilhos.setter
  def nFilhos(self, nFilhos):
    self._nFilhos = nFilhos
```

```python
class Arvore_Multi:

  def __init__(self):
    self.__raiz = None

  @property
  def raiz(self):
    return self.__raiz

  @raiz.setter
  def raiz(self, raiz):
    self.__raiz = raiz

  def busca_em_no(self, chave, no):
    n_Chaves = no.nFilhos - 1

    for i in range(n_Chaves):
      if (chave <= no.chaves[i].chave):
        return i

    return n_Chaves

  def busca_multi(self, chave):
    arvore = self.__raiz

    while (arvore != None):
      i = self.busca_em_no(chave, arvore)

      itemChave = arvore.chaves[i].chave

      if((i < arvore.nFilhos - 1) and (chave == itemChave)):
        return arvore.chaves[i].indice

      arvore = arvore.filhos[i]

    return -1

  def caminhamento(self, no):

    if (no != None):
      for i in range(no.nFilhos - 1):
        self.caminhamento(no.filhos[i])
        print(no.chaves[i].chave)

      self.caminhamento(no.filhos[no.nFilhos - 1])

  def set_filho(self, pai, posicao, no):
    pai.filhos[posicao] = no
```

```python
if __name__ == "__main__":

  arvore = Arvore_Multi()

  arvore.raiz = No([20, 30, 40])

  no1 = No([10, 15])
  no2 = No([25, 28])
  no3 = No([35, 38])
  no4 = No([50, 60, 70])

  arvore.set_filho(arvore.raiz, 0, no1)
  arvore.set_filho(arvore.raiz, 1, no2)
  arvore.set_filho(arvore.raiz, 2, no3)
  arvore.set_filho(arvore.raiz, 3, no4)

  arvore.caminhamento(arvore.raiz)
```
