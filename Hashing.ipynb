```python
class Computador():
  def __init__(self,ip,hardware):
    self.ip = ip
    self.hardware = hardware

  def __str__(self):
    return self.ip

  def __eq__(self, outro):
    return self.ip == outro.ip

  def __hash__(self):
    return hash(self.ip)


c1 = Computador('192.168.0.1','i7 8GB 1TB HD')
c2 = Computador('192.168.0.4','i5 8GB 1TB HD')

print(hash(c1))
print(hash(c2))

print(c1 == c2)
```
